```python
import pandas as pd
df = pd.read_excel("Country_Data_Final.xlsx")
```

```python
df.info()
```

```python
df.drop(columns=df.columns[df.columns.str.startswith("country_")], inplace=True)
df.drop(columns=df.columns[df.columns.str.startswith("bary_")], inplace=True)
df.drop(columns=["Capital Longitude", "Capital Latitude", "Area"], inplace=True)
```

```python
df.info()
```

```python
# Select columns 3-15 (inclusive, so up to index 15, meaning slice ends at 16)
columns_set1 = df.columns[3:16]
# Select columns 16-23 (inclusive, so up to index 23, meaning slice ends at 24)
columns_set2 = df.columns[16:24]

# Create an empty DataFrame to store the correlation results
correlation_matrix = pd.DataFrame(index=columns_set1, columns=columns_set2)

# Calculate the correlation between each column in set1 and each column in set2
for col1 in columns_set1:
    for col2 in columns_set2:
        correlation_matrix.loc[col1, col2] = df[col1].corr(df[col2])

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Define a significance threshold
significance_threshold = 0.1

# Filter the correlation matrix to show only significant correlations
significant_correlations = correlation_matrix[correlation_matrix.abs() > significance_threshold].dropna(how='all').dropna(axis=1, how='all')

# Display the significant correlations
print("Significant Correlations (absolute value > 0.5):")
print(significant_correlations)
```
